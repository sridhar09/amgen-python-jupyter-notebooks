Create a new column 'revenue' which is `item_cnt_day * item_price`

```python
sales['revenue'] = sales['item_cnt_day'] * sales['item_price']
sales.head()
```

What were the top 10 rows by revenue?

```python
series = sales.sort_values('revenue', ascending=False)
series.head(10)
```

```python
sales.sort_values('revenue', ascending=False).iloc[:10]
```

```python
sales.nlargest(10, 'revenue')
```

How many rows had more than 50 items (`item_cnt_day`) in them?

```python
(sales.item_cnt_day > 50).sum()
```
