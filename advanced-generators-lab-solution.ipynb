# Lab: Pipelines

Write a processing pipeline using generators (e.g. use the `def` and `yield` statements) that will print out the capitalized lines in a file that contain the word 'Hamlet' (any case).

(the file ./data/hamlet.txt would be good to test with)

```python
#gen_lines = open
def gen_lines(filename):
    """This is a 'source', so it does not take a generator as input."""
    with open(filename) as fp:
        for line in fp:
            yield line
            
def grep_for(gen, name):
    '''Your code here'''
    for line in gen:
        # if name.lower() in line.lower():
        if name.casefold() in line.casefold():
            yield line
    
def capitalize(gen):
    '''Your code here'''
    for line in gen:
        yield line.upper()
        # yield line.title()
```

```python
x = gen_lines('./data/hamlet.txt')
x = grep_for(x, 'hamlet')
x = capitalize(x)
```

```python
for line in x:
    print(line, end='')
```

```python
def capitalize_with_word(gen, name):
    for line in gen:
        if name.casefold() in line.casefold():
            yield line.upper()
        else:
            yield line
```

```python
x = gen_lines('./data/hamlet.txt')
x = capitalize_with_word(x, 'hamlet')
for line in x:
    print(line, end='')
```

# Lab: Coroutines

Write a processing pipeline that will:

- capitalize lines with 'hamlet'
- all-lowercase lines with 'horatio'
- omit all other lines

```python
def handle_lines(gen, upper_words, lower_words):
    for line in gen:
        for word in upper_words:
            if word.casefold() in line.casefold():
                yield line.upper()
                break
        for word in lower_words:
            if word.casefold() in line.casefold():
                yield line.lower()
                break
```

```python
x = gen_lines('./data/hamlet.txt')
x = handle_lines(x, ['hamlet'], ['horatio'])
for line in x:
    print(line, end='')
```

```python
def line_handler(lst):
    while True:
        line = yield
        lst.append(line)
    
def handle_horatio(line_handler):
    while True:
        line = yield
        if 'horatio' in line.casefold():
            line_handler.send(line.lower())

def handle_hamlet(line_handler):
    while True:
        line = yield
        if 'hamlet' in line.casefold():
            line_handler.send(line.upper())
```

```python
list_of_lines = []
lh = line_handler(list_of_lines)
next(lh)
handlers = [handle_horatio(lh), handle_hamlet(lh)]
```

```python
for coro in handlers:
    next(coro)
```

```python
with open('./data/hamlet.txt') as f:
    for line in f:
        for coro in handlers:
            coro.send(line)
```
